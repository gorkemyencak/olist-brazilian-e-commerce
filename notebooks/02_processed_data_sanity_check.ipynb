```python
# loading processed dataframes
loader = ProcessedLoader('olist')
data_raw = loader.load_all()

df_customers = data_raw["processed_customers_dataset.csv"]
df_geolocation = data_raw["processed_geolocation_dataset.csv"]
df_items = data_raw["processed_items_dataset.csv"]
df_payments = data_raw["processed_payments_dataset.csv"]
df_reviews = data_raw["processed_reviews_dataset.csv"]                     
df_orders = data_raw["processed_orders_dataset.csv"]
df_products = data_raw["processed_products_dataset.csv"]
df_sellers = data_raw["processed_sellers_dataset.csv"]
df_product_category = data_raw["processed_product_category_dataset.csv"]
```

```python
# summary of datasets
def summary(
        df,
        name
):
    print(f"-- {name.upper()} Dataset --")
    display(df.head())
    print(df.info())
    print(df.shape)
    print(df.isna().sum().sort_values(ascending=False).head(10))
```
